# Turistas no Rio de Janeiro

Esse dataset contém o número de todos os turistas no Rio de Janeiro entre 2006 e 2019.

Keaggle: https://www.kaggle.com/datasets/victorsoeiro/tourist-arrivals-in-rio-de-janeiro-20062019

```python
import pandas as pd 
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('tourists-rj-2006-2019.csv')
display(df.head())
```

# Considerações levadas por esse gráfico:

• É de se esperar que em 2014 houvessem muitos turistas no Rio de Janeiro, por conta da Copa do Mundo.

```python
graph = px.histogram(df, x='Ano', y='Total', title='Núm. de turistas por ano')
graph.update_layout(bargap=0.2)
graph.show()
```

# Considerações levadas por esse gráfico:

• Em 2016, os Estados Unidos eram o país que mais recebia turistas no Rio de Janeiro, porém, foi ultrapassado em 2009 pela Argentina

• Em 2019 os paises que o Rio de Janeiro mais recebeu turistas foram:

    1°: Argentina

    2°: Chile
    
    3°: Estados Unidos

```python
graph = px.line(df, x='Ano', y='Total', color='País', text='Ano', title='Países de onde o RJ recebeu mais turistas')
graph.show()
```

# Considerações levadas por esse gráfico:

• Recebemos muitos turistas pelas vias aéreas.

```python
total_maritima = df['Marítima'].sum()
total_aerea = df['Aérea'].sum()

graph = px.pie(df, values=[total_aerea, total_maritima], names=['Aérea', 'Marítima'], title='Meios de locomoção', 
               width=500, height=400)
graph.show()
```

# Considerações levadas por esse gráfico:

• O continente Europeu é o continente de onde mais o RJ recebe turistas.

```python
graph = px.histogram(df, x='Continente', title='Continentes')
graph.update_xaxes(categoryorder='total descending')
graph.show()
```
